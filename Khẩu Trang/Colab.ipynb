# Chào mừng bạn đến với Colab!

```
import matplotlib.pyplot as plt
import numpy as np

epochs = np.arange(1, 51)

train_loss = np.exp(-epochs/15) + 0.05
val_loss = np.exp(-epochs/14) + 0.08

precision = 1 - train_loss
recall = 1 - val_loss

plt.figure(figsize=(8,5))

plt.plot(epochs, train_loss, label='Train Loss')
plt.plot(epochs, val_loss, label='Validation Loss')

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Learning Curve")
plt.legend()
plt.grid(True)

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

cm = np.array([[280, 20],
               [18, 282]])

labels = ["Mask", "No Mask"]

plt.figure(figsize=(6,5))

sns.heatmap(cm,
            annot=True,
            fmt="d",
            cmap="Blues",
            xticklabels=labels,
            yticklabels=labels)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")

plt.show()
```

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import matplotlib.pyplot as plt

# Nhập số lượng từ bàn phím
mask = int(input("Nhập số lượng ảnh Mask: "))
nomask = int(input("Nhập số lượng ảnh No Mask: "))

classes = ['Mask', 'No Mask']
counts = [mask, nomask]

plt.figure(figsize=(8,6))

bars = plt.bar(classes, counts)

# Hiển thị số trên đầu cột
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval,
             int(yval), ha='center', va='bottom', fontsize=12)

plt.title("Số lượng đối tượng theo lớp", fontsize=16)
plt.xlabel("Lớp", fontsize=14)
plt.ylabel("Số lượng", fontsize=14)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```
from google.colab import drive
drive.mount('/content/drive')

import os
import matplotlib.pyplot as plt

base_path = "/content/drive/MyDrive/mask_new/dataset"

splits = ["train", "valid", "test"]

small = 0
medium = 0
large = 0

for split in splits:
    label_path = os.path.join(base_path, split, "labels")

    for file in os.listdir(label_path):
        if file.endswith(".txt"):
            with open(os.path.join(label_path, file), 'r') as f:
                for line in f:
                    parts = line.strip().split()
                    width = float(parts[3])
                    height = float(parts[4])
                    area = width * height

                    if area < 0.02:
                        small += 1
                    elif area < 0.15:
                        medium += 1
                    else:
                        large += 1

print("Small:", small)
print("Medium:", medium)
print("Large:", large)

# Vẽ biểu đồ
sizes = ['Kích thước nhỏ', 'Kích thước vừa', 'Kích thước lớn']
counts = [small, medium, large]

plt.figure(figsize=(8,6))
bars = plt.bar(sizes, counts)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval,
             int(yval), ha='center', va='bottom', fontsize=12)

plt.title("Phân bố kích thước đối tượng", fontsize=16)
plt.xlabel("Loại kích thước", fontsize=14)
plt.ylabel("Số lượng", fontsize=14)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# BCE loss
def bce(y, p):
    return -(y*np.log(p) + (1-y)*np.log(1-p))

p = np.linspace(0.001, 0.999, 200)

loss_mask = bce(1, p)
loss_nomask = bce(0, p)

plt.figure(figsize=(8,6))
plt.plot(p, loss_mask, label="Mask (y=1)")
plt.plot(p, loss_nomask, label="No Mask (y=0)")

plt.xlabel("Xác suất dự đoán")
plt.ylabel("Loss")
plt.title("Binary Cross Entropy Loss cho bài toán phát hiện khẩu trang")
plt.legend()
plt.grid(True)
plt.show()
```

```
iou = np.linspace(0, 1, 200)
loss = 1 - iou

plt.figure(figsize=(8,6))
plt.plot(iou, loss)

plt.xlabel("IoU")
plt.ylabel("Box Loss")
plt.title("Localization Loss (1 - IoU)")
plt.grid(True)
plt.show()
```

```
obj = np.linspace(0.001, 0.999, 200)
loss_obj = bce(1, obj)

plt.figure(figsize=(8,6))
plt.plot(obj, loss_obj)

plt.xlabel("Độ tin cậy dự đoán")
plt.ylabel("Loss")
plt.title("Objectness Loss trong YOLO")
plt.grid(True)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# 1. Hàm mất mát giả lập
def f(x, y):
    return x**2 + y**2

# 2. Gradient của hàm
def grad(x, y):
    return np.array([2*x, 2*y])

# 3. Gradient Descent
def gradient_descent(lr=0.1, steps=20):
    x, y = 3, 3   # điểm bắt đầu
    path = []

    for _ in range(steps):
        path.append((x, y))
        g = grad(x, y)
        x -= lr * g[0]
        y -= lr * g[1]

    return np.array(path)

# 4. Tạo contour của hàm loss
x = np.linspace(-4,4,100)
y = np.linspace(-4,4,100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# 5. Vẽ biểu đồ
plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, 30)

gd_path = gradient_descent()

plt.plot(gd_path[:,0], gd_path[:,1], 'o-', color='red', label='Gradient Descent')

plt.title("Gradient Descent Optimization")
plt.xlabel("Weight 1")
plt.ylabel("Weight 2")
plt.legend()
plt.grid(True)
plt.show()
```
